```python
# Check if a Number is Positive, Negative or Zero

a=int(input("type the no."))
if a>0:
    print("the no is positive")
elif a<0:
    print("the no is negative")
else:
    print("the no is zero")
```

```python
# if a Number is Odd or Even

x=int(input("type the no."))
if x%2==0 :
    print("the no is Even")
else:
    print("the no is Odd")
```

```python
# to Check Leap Year
y=int(input("type the given year"))
if y%4==0 :
    print("It's a Leap Year")
else:
    print("It's not a Leap Year")
```

```python
# to Check Prime Number

a=int(input("enter the no."))
if(a>1):
    for i in range(2,a):
        if(a%i==0):
            print("it's not a prime no")
            break
    else:
        print("it's a prime no")
        
else:
    print("it's not a prime no")
```
